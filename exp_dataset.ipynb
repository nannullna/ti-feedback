```python
import os
import sys
import json
from glob import glob

import numpy as np
import pandas as pd
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from transformers import CLIPConfig, CLIPImageProcessor, CLIPVisionModel, CLIPTokenizer

from tqdm import tqdm

MODEL_NAME = "openai/clip-vit-large-patch14"
```

```python
class CLIPImageDataset(Dataset):

    def __init__(self, image_dir, processor: CLIPImageProcessor):
        super().__init__()
        self.image_dir = image_dir
        self.processor = processor
        self.image_paths = glob(os.path.join(image_dir, "*.png"))
        self.image_paths = sorted(self.image_paths)

    def __len__(self):
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image = Image.open(image_path)
        image = self.processor(image=image, return_tensors="pt")["pixel_values"]
        return image
```

```python
config = CLIPConfig.from_pretrained(MODEL_NAME)
processor = CLIPImageProcessor.from_pretrained(MODEL_NAME)
tokenizer = CLIPTokenizer.from_pretrained(MODEL_NAME)
model = CLIPVisionModel.from_pretrained(MODEL_NAME, config=config)
```
